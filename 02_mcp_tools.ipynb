# 02. MAF 用ツールを作成する (MCP ツール)

## コードの解説

**MCP (Model Context Protocol)** は AI エージェントが外部ツールを呼び出すための標準プロトコルです。
MAF の `MCPStreamableHTTPTool` を使うと、MCP サーバーが提供するツール群を一括でエージェントに組み込めます。

### カスタムツールとの違い

| 比較項目 | カスタムツール | MCP ツール |
|--------|-------------|----------|
| 定義場所 | Python コード内で直接定義 | 外部 MCP サーバー上で定義 |
| ツール数 | 1 関数 = 1 ツール | 1 オブジェクトでサーバー上の全ツールを利用可能 |
| 実行場所 | ローカル Python プロセス | 外部 HTTP サーバー |
| 用途 | 簡単なロジック、プロトタイプ | 外部システム統合、既存サービスの活用 |

### MCPStreamableHTTPTool の設定パラメータ

| パラメータ | 説明 |
|---------|------|
| `name` | ツールグループの識別名 |
| `url` | MCP サーバーのエンドポイント URL |
| `description` | エージェントがツールを選択する際の説明文 |
| `timeout` | 接続タイムアウト（秒）。初回接続は長めに設定推奨 |

### MCP サーバーの起動方法

このワークショップでは `mcp/mock_server.py` にモック MCP サーバーが含まれています。
ターミナルで以下のコマンドを実行して起動してください：

```bash
python mcp/mock_server.py
```

起動後、`MCP_SERVER_URI` 環境変数が指すエンドポイントでツールが利用可能になります。

````python
import os
from dotenv import load_dotenv
from agent_framework import MCPStreamableHTTPTool

load_dotenv()

# 環境変数から MCP サーバーの URL を取得
# .env ファイルに MCP_SERVER_URI=http://localhost:8000/mcp のように設定する
MCP_SERVER_URI = os.getenv("MCP_SERVER_URI", "http://localhost:8000/mcp")


# ---------------------------------------------------------------
# MCPStreamableHTTPTool の作成
# MCPStreamableHTTPTool は MCP サーバー上のツール群を
# 一括でエージェントに提供するための MAF のツールクラス
#
# ポイント1: url に MCP サーバーのエンドポイントを指定する
# ポイント2: description はエージェントがツール選択する際の判断材料になる
# ポイント3: timeout は長めに設定（初回接続が重い場合がある）
# ---------------------------------------------------------------
mcp_tool = MCPStreamableHTTPTool(
    name="game_shop_tools",
    url=MCP_SERVER_URI,
    description=(
        "Xbox ゲームショップの製品情報、注文管理、在庫確認、"
        "ユーザー情報、CRM/SNS 分析データを取得するツール群です。"
    ),
    timeout=120,
)

print("MCP ツール作成完了")
print(f"  名前  : {mcp_tool.name}")
print(f"  接続先: {MCP_SERVER_URI}")
print()
print("次のステップ: このツールを Agent に渡すと、")
print("MCP サーバーが提供するすべてのツールが自動的に利用可能になります。")
````
